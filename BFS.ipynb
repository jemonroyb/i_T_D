**Busqueda en amplitud**

**Ejemplo 1**

graf3.png

```python
graph = {
  '0' : ['3','4'],
  '3' : ['0', '1'],
  '1' : ['3','5'],
  '4' : ['0','2','5'],
  '5' : ['4','1'],
  '2' : ['4']
  
}

visited = [] # Lista de nodos visitados.
queue = []     #Inicializar una cola

def bfs(visited, graph, node): #Funcin para BFS
  visited.append(node)
  queue.append(node)

  while queue:          # Creacin de bucle para visitar cada nodo
    m = queue.pop(0) 
    print (m, end = " ") 

    for neighbour in graph[m]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)

# Nodo inicio
print("Bsqueda en amplitud")
bfs(visited, graph, '0')    # Llamada de funcin
```
